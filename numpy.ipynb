# NumPy: cours OpenClassrooms

NumPy est une lib qui permet d'effectuer rapidement et simplement des calculs complexes grâce aux Arrays.

```python
import numpy as np
```

```python
revenus = [1400, 1700, 2100, 3000, 6000, 550]

print("Moyenne des revenus : " + str(round(np.mean(revenus),2)) + " €")
```

`np.mean()` est une fonction de NumPy qui permet de calculer la moyenne.
Elle accepte aussi bien un array NumPy qu’une liste Python (ou tout objet array-like).

`.mean()` est une méthode de la classe ndarray (l’objet principal de NumPy). Elle ne peut être utilisée que sur un array NumPy.

```python
revenus_moy = np.array(revenus).mean()

print(round(revenus_moy, 2))
```

```python
print(np.zeros(10))
print(np.ones(3))
print(np.arange(2, 12, 2))
print(np.linspace(1, 60, 5))
print(np.power(2, np.arange(1, 14)))
```
